```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# a) Vectores
a = np.array([3.1,1,-0.5,-3.2,6])
b = np.array([1,3,2.2,5.1,1])

# b) Multiplicacion escalar
m_e = a @ b
print("Multiplicacion escalar")
print(m_e)

# c) Multiplicacion punto a punto
m_p = a * b
print("\nMultiplicacion punto a punto")
print(m_p)

# d) Matriz

A = np.array([[2,-1,-3],[4,1.5,-2.5],[7.3,-0.9,0.2]])
print("\nMatriz A")
print(A)

# e) Traspuesta

tras_A = A.T
print("\nTraspuesta de la matriz A")
print(tras_A)

# f) Comandos ones, round, ceil, floor

matriz_unos = np.ones((3, 3))
print("\nComando ones")
print(matriz_unos)

arr = np.array([1.2, 2.5, 3.7])
redondeado = np.round(arr)
print("\nComando round")
print(redondeado)

r_arriba = np.ceil(arr)
print("\nComando ceil")
print(r_arriba)

r_abajo = np.floor(arr)
print("\nComando floor")
print(r_abajo)

# g) Valor matriz

valor = A[0,2]
print("\nElemento primera fila tercera columna de la matriz A")
print(valor)

# h) Fila matriz

valor1 = A[1,:]
print("\nSegunda fila de la matriz A")
print(valor1)

# i) Dimensiones de matriz

dim = A.shape
print("\nDimensiones de la matriz A")
print(dim)

# j) Funcion y(n) = sin(pi*0.12*n) en el intervalo 0<=n>=100

n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)

# k) Funcion y2(n) = cos(2*pi*0.03*n) en el intervalo 0<=n>=100

y2 = np.cos(2 * np.pi * 0.03 * n)

# l) Funcion s(n) = y(n) + y2(n), Funcion t(n) = y(n) * y2(n)

s = y + y2
t = y * y2

# m) Grafica de la señal y(n) y y2(n)

plt.plot(y, label='y(n)')
plt.plot(y2, label='y2(n)')
plt.legend()
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señal y(n) y y2(n)')
plt.show()

print("\n")
# n) Grafica de la señal s(n) y t(n)

plt.plot(s, label='s(n)')
plt.plot(t, label='t(n)')
plt.legend()
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señal s(n) y t(n)')
```

```python
# PANDAS
# 1) Funcion diccionario

def notas(notas_dict):
    notas = np.array(list(notas_dict.values()))

    # Calculamos los valores estadísticos
    resumen = pd.Series({
        'Nota mínima': np.min(notas),
        'Nota máxima': np.max(notas),
        'Nota media': np.mean(notas),
        'Desviación estándar': np.std(notas, ddof=1)  # ddof=1 para muestra
    })

    return resumen

# Ejemplo de uso
notas_alumnos = {'Ana': 4.5, 'Juan': 3.8, 'Carlos': 4.2, 'Elena': 4.9, 'Luis': 3.5}
resultado = notas(notas_alumnos)
print(resultado)

```

```python
# 2) Calculo de indice de masa corporal IMC

# a) Cargar datos

datos = pd.read_csv('datos.csv', sep=';')

# b) Mostrar primeras y ultimas filas

print(datos.head())
print(datos.tail())

# c) Borrar colomna Unnamed: 0

datos = datos.drop('Unnamed: 0', axis=1)
print("\n")
print(datos.head())

# d) Calcular IMC

peso = datos.Weight
altura = datos.Height

imc = peso/((altura/100)**2)

# e) Agregar IMC como una nueva columna llamada BMI

datos['BMI'] = imc

# f) Mostrar primeras y ultimas filas

print("\n")
print(datos.head())
print(datos.tail())

# e) Categorizar segun IMC en Bajo peso (<18.5), Normal (18.5-24.9), sobrepeso (25 - 29.9), obesidad (>=30)

datos['Categoria_IMC'] = pd.cut(datos['BMI'],
                             bins=[0, 18.5, 24.9, 29.9, float('inf')],
                             labels=['Bajo peso', 'Normal', 'Sobrepeso', 'Obesidad'])

print(datos[['BMI', 'Categoria_IMC']].head())  # Verifica las categorías
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función w(t)
def w(t):
    if -2 <= t < -1:
        return t + 1.5
    elif -1 <= t < 1:
        return 0
    elif 1 <= t < 2:
        return 2*t + 3
    elif 2 <= t < 3:
        return -2*t - 3
    else:
        return 0

# Generar valores de t
t_values = np.linspace(-3, 4, 1000)
w_values = np.array([w(t) for t in t_values])

# Graficar la función w(t)
plt.figure(figsize=(8, 5))
plt.plot(t_values, w_values, label=r'$w(t)$', color='b')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid()
plt.xlabel("t")
plt.ylabel("w(t)")
plt.title("Gráfica de la señal w(t)")
plt.legend()
plt.show()
```
