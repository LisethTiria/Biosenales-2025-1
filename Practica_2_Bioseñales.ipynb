```python
import numpy as np
import matplotlib.pyplot as plt
```

**Impulso unitario**

```python
def impseq(n0, n1, n2):

#   Genera x(n) = delta(n-n0); n1 <= n <= n2

#	 ----------------------------------------------

    n = np.arange(n1,n2+1) # Se crea el vector de muestras

    x = ((n-n0) == 0).astype(int)

    return [x,n]
x, n = impseq(0, -3, 3)
print("n:", n)
print("x:", x)
```

**Escalon**

```python
#función escalón unitario, definida en un intervalo n1 ≤ n0 ≤ n2

def escalon_unitario(n0, n1, n2):
    # Genera x(n) = u(n - n0); n1 <= n <= n2

    # Se crea el vector de muestras n, que va desde n1 hasta n2
    n = np.arange(n1, n2 + 1)

    # Se genera la secuencia escalón: es 1 cuando n >= n0, y 0 en otro caso
    x = (n >= n0).astype(int)

    return x, n  # Se retorna la señal escalón y los valores de n

x, n = escalon_unitario(0, -3, 3)
print("n:", n)
print("x:", x)
```

**Rampa**

```python
# función rampa, definida en un intervalo n1 ≤ n0 ≤ n2 con una pendiente m.

def rampa_unitaria(n0, n1, n2, m):
    # Genera x(n) = m * (n - n0) para n >= n0, 0 en otro caso
    # n1 <= n <= n2

    # Se crea el vector de muestras n, que va desde n1 hasta n2
    n = np.arange(n1, n2 + 1)

    # Se genera la secuencia rampa: m * (n - n0) si n >= n0, de lo contrario 0
    x = m * (n - n0) * (n >= n0)

    return x, n  # Se retorna la señal rampa y los valores de n

x, n = rampa_unitaria(0, -3, 3,2)
print("n:", n)
print("x:", x)

```
